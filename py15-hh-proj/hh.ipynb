```python

def get_graids(text):
    if text.startswith('Неоконч'):
        return 'неоконченное высшее'
    elif text.startswith('Высш'):
        return 'высшее'
    elif text.startswith('Среднее спец'):
        return 'среднее специальное'
    else:
        return 'среднее'
    
hh_db_t['graid'] = hh_db_t.iloc[:,-3].apply(get_graids)
hh_db_t.head()
```

```python
hh_db_t['graid'].value_counts()
```

```python
# пол
def get_gender(text):
    if text.startswith('Муж'):
        return 'М'
    elif text.startswith('Жен'):
        return 'Ж'
hh_db_t['gender'] = hh_db_t.iloc[:,0].apply(get_gender)
hh_db_t['gender'].value_counts(normalize=True)
```

```python
# возраст
def get_age(text):
    text = text.split(',')
    text = text[1]
    text = text.split(' ')
    text = text[2]
    return int(text)
    
hh_db_t['age'] = hh_db_t.iloc[:,0].apply(get_age)
hh_db_t['age'].head()
```

```python
hh_db_t['age'].mean()
```

```python
hh_db_t = hh_db_t.drop(['Образование и ВУЗ'], axis=1)
hh_db_t = hh_db_t.drop(['Пол, возраст'], axis=1)
hh_db_t.head()
```

```python
# опыт
def get_exp(text1):
    month_count = 0
    if text1 is np.NaN or text1 == 'Не указано':
        return np.NaN
    year_words=['год', 'года', 'лет']
    month_words=['месяц', 'месяца', 'месяцев']
    arg_splitted = text1.split(' ')[:7]
    years = 0
    months = 0
    for index, item in enumerate (arg_splitted):
        if item in year_words:
            years = int(arg_splitted[index-1])
        if item in month_words:
            months = int(arg_splitted[index-1])
    return int(years*12 + months)
hh_db_t['exp'] = hh_db_t.iloc[:, 5].apply(get_exp)
```

```python
round(hh_db_t['exp'].median())
```

```python
# Город, переезд, командировки
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]

def get_city(str_city):           
    str_city = str_city.replace(' ', '')
    str_city = str_city.split(',')
    
    if str_city[0] == 'Москва':
        city = 'Москва'
    elif str_city[0] == 'Санкт-Петербург':
        city = 'Санкт-Петербург'
    elif str_city[0] in million_cities:
        city = 'город-миллионник'
    else:
        city = 'другие'
        
    return city


def get_moving(moving):
    if '(' in moving:
        brackets = moving[moving.find('('):moving.find(')')]
        moving = moving.replace(brackets,'')
        
    moving = moving.replace(' ', '')
    moving = moving.split(',')
            
    # check station
    if len(moving) <2:
        moving = False
    elif len(moving) == 2 and moving[1].startswith('м.'):
        moving = False   
            
    else:
        if moving[1].startswith('м.'):
            moving.pop(1)
        if moving[1].startswith('негот'):
                moving = False                
        elif moving[1].startswith('гот') or moving[1].startswith('хочу'):
                moving = True 
    return moving


def get_trip(trip):
        
    trip = trip.replace(' ', '')
    trip = trip.split(',')
            
    # check station
    if len(trip) <3:
        trip = False
    elif len(trip) <= 3 and trip[1].startswith('м.'):
        trip = False   
            
    else:
        
        if trip[-1].startswith('негот'):
                trip = False                
        elif trip[-1].startswith('гот'):
                trip = True 
        else:
            trip = False
    return trip
            
            
hh_db_t['Город'] = hh_db_t.iloc[:, 2].apply(get_city)
hh_db_t['Готовность к переезду'] =  hh_db_t.iloc[:, 2].apply(get_moving)
hh_db_t['Готовность к командировкам']= hh_db_t.iloc[:, 2].apply(get_trip)
hh_db_t.head(10)
```
